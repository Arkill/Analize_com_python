```python
# quantos exemplares foram emprestados por ano 

emprestimos_data = pd.DataFrame(emprestimos_completo['data_emprestimo'].value_counts()).reset_index()
emprestimos_data.columns = ['data','quantidade']
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])
emprestimos_data
```

```python
# Exibir os nomes das colunas para verificar
print(emprestimos_data.columns)
```

```python
# Converta 'data' para datetime, se necessário
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])

# Agrupar por ano e somar os valores de 'valor'
emprestimos_por_ano = emprestimos_data.groupby(emprestimos_data['data'].dt.year)['quantidade'].sum()

# Renomear o índice para 'ano'
emprestimos_por_ano.index.name = 'ano'

# Exibir o resultado
print(emprestimos_por_ano)
```

```python
# Transforme a série em um DataFrame com reset_index()
emprestimos_por_ano_df = emprestimos_por_ano.reset_index()
```

```python
# Configuração do tema dos gráficos
sns.set_theme(context='notebook', 
              style='darkgrid', # Alterado para 'dark' para obter fundo preto
              palette='deep', 
              font_scale=1.3, 
              rc={"figure.figsize":(20,8), 'axes.facecolor': 'black'})  # Configuração do fundo preto

# Plotagem do gráfico de linha
ax = sns.lineplot(data=emprestimos_por_ano_df, x='ano', y='quantidade', color='red')

# Configurações adicionais do eixo x
ax.set(xlabel=None, ylabel=None)
ax.tick_params(axis='x', rotation=30)

# Formatação do eixo y
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',').replace(',', '.')))

# Configuração do título
ax.set_title('Quantidade de exemplares emprestados do SISBI por ano'+'\n', size=20, loc='left', weight='bold')


# Exibir o gráfico
plt.show()
```

```python
# emprestimo por cada mês

# Supondo que 'emprestimos_data' é seu DataFrame e que a coluna de datas se chama 'data'
# Certifique-se de que a coluna 'data' está no formato datetime
emprestimos_data['data'] = pd.to_datetime(emprestimos_data['data'])

# Agrupe os dados por mês e some apenas as colunas numéricas
emprestimos_por_mes = emprestimos_data.groupby(emprestimos_data['data'].dt.month).sum(numeric_only=True)

# Renomeie o índice para 'mes'
emprestimos_por_mes.index.name = 'mes'

# Exiba o resultado
print(emprestimos_por_mes)
```
